## Perhitungan Manual 

Berdasarkan soal UTS dan konfirmasi dari dosen, nilai input (crisp) yang dipilih adalah:
* **Permintaan (X) = 2069** 
* **Persediaan (Y) = 569** 

### Step 1: Fuzzifikasi (Mencari nilai keanggotaan μ)

Mencari derajat keanggotaan (μ) dari setiap input berdasarkan grafik `permintaan.png` dan `persediaan.png`.

**1. Variabel Permintaan (X = 2069)**
* **μ Turun(2069):** (Grafik linear turun [1000, 1] ke [3000, 0])
    * Rumus: `(Batas Atas - X) / (Batas Atas - Batas Bawah)`
    * μ\_turun = (3000 - 2069) / (3000 - 1000)
    * μ\_turun = 931 / 2000 = **0.4655**
* **μ Naik(2069):** (Grafik linear naik [1000, 0] ke [3000, 1])
    * Rumus: `(X - Batas Bawah) / (Batas Atas - Batas Bawah)`
    * μ\_naik = (2069 - 1000) / (3000 - 1000)
    * μ\_naik = 1069 / 2000 = **0.5345**

**2. Variabel Persediaan (Y = 569)**
* **μ Sedikit(569):** (Grafik linear turun [200, 1] ke [400, 0])
    * Karena 569 berada di luar rentang (≥ 400), nilainya **0**.
* **μ Sedang(569):** (Grafik segitiga [200, 0] -> [400, 1] -> [800, 0]. Nilai 569 ada di sisi turun [400, 1] ke [800, 0])
    * Rumus: `(Batas Atas - Y) / (Batas Atas - Batas Tengah)`
    * μ\_sedang = (800 - 569) / (800 - 400)
    * μ\_sedang = 231 / 400 = **0.5775**
* **μ Banyak(569):** (Grafik linear naik [400, 0] ke [800, 1])
    * Rumus: `(Y - Batas Bawah) / (Batas Atas - Batas Bawah)`
    * μ\_banyak = (569 - 400) / (800 - 400)
    * μ\_banyak = 169 / 400 = **0.4225**

### Step 2: Inferensi (Evaluasi Aturan)

Mencari nilai predikat (alpha, α) menggunakan operator `AND (MIN)` dan mencari nilai `z` (produksi) untuk setiap aturan yang aktif (α > 0) berdasarkan grafik `produksi.png`.

* Rumus `z BERKURANG`: `z = 7000 - (α * (7000 - 2000))` -> **`z = 7000 - (α * 5000)`**
* Rumus `z BERTAMBAH`: `z = (α * (7000 - 2000)) + 2000` -> **`z = (α * 5000) + 2000`**

* **[R1]** α1 = min(TURUN: 0.4655, BANYAK: 0.4225) = **0.4225**
    * z1 (BERKURANG) = 7000 - (0.4225 * 5000) = **4887.5**
* **[R2]** α2 = min(TURUN: 0.4655, SEDANG: 0.5775) = **0.4655**
    * z2 (BERKURANG) = 7000 - (0.4655 * 5000) = **4672.5**
* **[R3]** α3 = min(TURUN: 0.4655, SEDIKIT: 0) = **0**
* **[R4]** α4 = min(NAIK: 0.5345, BANYAK: 0.4225) = **0.4225**
    * z4 (BERKURANG) = 7000 - (0.4225 * 5000) = **4887.5**
* **[R5]** α5 = min(NAIK: 0.5345, SEDANG: 0.5775) = **0.5345**
    * z5 (BERTAMBAH) = (0.5345 * 5000) + 2000 = **4672.5**
* **[R6]** α6 = min(NAIK: 0.5345, SEDIKIT: 0) = **0**

### Step 3: Defuzzifikasi (Weighted Average)

Menghitung nilai Z akhir (jumlah produksi) menggunakan metode rata-rata tertimbang (Weighted Average).

* **Pembilang (Total α * z):**
    * (α1 * z1) = (0.4225 * 4887.5) = 2063.84375
    * (α2 * z2) = (0.4655 * 4672.5) = 2174.90125
    * (α4 * z4) = (0.4225 * 4887.5) = 2063.84375
    * (α5 * z5) = (0.5345 * 4672.5) = 2496.79125
    * **Total Pembilang = 8802.4375**
* **Penyebut (Total α):**
    * α1 + α2 + α4 + α5 = 0.4225 + 0.4655 + 0.4225 + 0.5345
    * **Total Penyebut = 1.845**
* **Z (Hasil Akhir):**
    * Z = 8802.4375 / 1.845 = **4770.9688...**
    * Dibulatkan: **4771**

### Kesimpulan

Jadi, jumlah kemasan makanan jenis ABC yang harus diproduksi, dengan permintaan **2069** dan persediaan **569**, adalah **4771 kemasan** (dibulatkan).

## Source code

```python
# Cell untuk import library
import matplotlib.pyplot as plt
import numpy as np
```

```python
# -------------------------------------
# FUNGSI FUZZIFIKASI
# -------------------------------------

def fuzzify_permintaan(x):
    """
    Menghitung derajat keanggotaan (μ) untuk variabel Permintaan.
    Berdasarkan grafik permintaan.png
    """
    mu_turun = 0.0
    mu_naik = 0.0

    # μ Turun (Linear turun: (1000, 1) -> (3000, 0))
    if x <= 1000:
        mu_turun = 1.0
    elif x > 1000 and x < 3000:
        mu_turun = (3000.0 - x) / (3000.0 - 1000.0)
    else: # x >= 3000
        mu_turun = 0.0

    # μ Naik (Linear naik: (1000, 0) -> (3000, 1))
    if x <= 1000:
        mu_naik = 0.0
    elif x > 1000 and x < 3000:
        mu_naik = (x - 1000.0) / (3000.0 - 1000.0)
    else: # x >= 3000
        mu_naik = 1.0
        
    return {'turun': mu_turun, 'naik': mu_naik}

def fuzzify_persediaan(y):
    """
    Menghitung derajat keanggotaan (μ) untuk variabel Persediaan.
    Berdasarkan grafik persediaan.png
    """
    mu_sedikit = 0.0
    mu_sedang = 0.0
    mu_banyak = 0.0

    # μ Sedikit (Linear turun: (200, 1) -> (400, 0))
    if y <= 200:
        mu_sedikit = 1.0
    elif y > 200 and y < 400:
        mu_sedikit = (400.0 - y) / (400.0 - 200.0)
    else: # y >= 400
        mu_sedikit = 0.0

    # μ Sedang (Segitiga: (200, 0) -> (400, 1) -> (800, 0))
    if y <= 200 or y >= 800:
        mu_sedang = 0.0
    elif y > 200 and y <= 400:
        mu_sedang = (y - 200.0) / (400.0 - 200.0)
    elif y > 400 and y < 800:
        mu_sedang = (800.0 - y) / (800.0 - 400.0)

    # μ Banyak (Linear naik: (400, 0) -> (800, 1))
    if y <= 400:
        mu_banyak = 0.0
    elif y > 400 and y < 800:
        mu_banyak = (y - 400.0) / (800.0 - 400.0)
    else: # y >= 800
        mu_banyak = 1.0
        
    return {'sedikit': mu_sedikit, 'sedang': mu_sedang, 'banyak': mu_banyak}

# -------------------------------------
# FUNGSI INFERENSI (TSUKAMOTO)
# -------------------------------------

def get_z_produksi(alpha, status):
    """
    Mencari nilai crisp 'z' dari output fuzzy (Produksi).
    Berdasarkan grafik produksi.png (Range 2000-7000, selisih 5000)
    """
    if status == 'BERKURANG':
        # z = 7000 - (alpha * 5000)
        return 7000.0 - (alpha * (7000.0 - 2000.0))
    elif status == 'BERTAMBAH':
        # z = (alpha * 5000) + 2000
        return (alpha * (7000.0 - 2000.0)) + 2000.0
    return 0.0 # Harusnya tidak terjadi

# -------------------------------------
# PROGRAM UTAMA
# -------------------------------------

# Input Crisp sesuai pilihan saya   
permintaan_in = 2069
persediaan_in = 569

print(f"Input Crisp: Permintaan = {permintaan_in}, Persediaan = {persediaan_in}\n")

# ----- Step 1: Fuzzifikasi -----
mu_permintaan = fuzzify_permintaan(permintaan_in)
mu_persediaan = fuzzify_persediaan(persediaan_in)

print("--- Hasil Fuzzifikasi ---")
print(f"Permintaan: {mu_permintaan}")
print(f"Persediaan: {mu_persediaan}\n")

# ----- Step 2: Inferensi (Evaluasi Aturan) -----
alpha = {}
z = {}

# [R1] IF Permintaan TURUN AND Persediaan BANYAK THEN Produksi BERKURANG
alpha[1] = min(mu_permintaan['turun'], mu_persediaan['banyak'])
z[1] = get_z_produksi(alpha[1], 'BERKURANG')

# [R2] IF Permintaan TURUN AND Persediaan SEDANG THEN Produksi BERKURANG
alpha[2] = min(mu_permintaan['turun'], mu_persediaan['sedang'])
z[2] = get_z_produksi(alpha[2], 'BERKURANG')

# [R3] IF Permintaan TURUN AND Persediaan SEDIKIT THEN Produksi BERTAMBAH
alpha[3] = min(mu_permintaan['turun'], mu_persediaan['sedikit'])
z[3] = get_z_produksi(alpha[3], 'BERTAMBAH')

# [R4] IF Permintaan NAIK AND Persediaan BANYAK THEN Produksi BERKURANG
alpha[4] = min(mu_permintaan['naik'], mu_persediaan['banyak'])
z[4] = get_z_produksi(alpha[4], 'BERKURANG')

# [R5] IF Permintaan NAIK AND Persediaan SEDANG THEN Produksi BERTAMBAH
alpha[5] = min(mu_permintaan['naik'], mu_persediaan['sedang'])
z[5] = get_z_produksi(alpha[5], 'BERTAMBAH')

# [R6] IF Permintaan NAIK AND Persediaan SEDIKIT THEN Produksi BERTAMBAH
alpha[6] = min(mu_permintaan['naik'], mu_persediaan['sedikit'])
z[6] = get_z_produksi(alpha[6], 'BERTAMBAH')

print("--- Hasil Inferensi (Tsukamoto) ---")
print(f"[R1] alpha = {alpha[1]:.4f} -> z = {z[1]:.4f} (Berkurang)")
print(f"[R2] alpha = {alpha[2]:.4f} -> z = {z[2]:.4f} (Berkurang)")
print(f"[R3] alpha = {alpha[3]:.4f} -> z = {z[3]:.4f} (Bertambah)")
print(f"[R4] alpha = {alpha[4]:.4f} -> z = {z[4]:.4f} (Berkurang)")
print(f"[R5] alpha = {alpha[5]:.4f} -> z = {z[5]:.4f} (Bertambah)")
print(f"[R6] alpha = {alpha[6]:.4f} -> z = {z[6]:.4f} (Bertambah)\n")

# ----- Step 3: Defuzzifikasi (Weighted Average) -----
numerator = 0.0   # Pembilang: (alpha[i] * z[i])
denominator = 0.0 # Penyebut: (alpha[i])

# Loop ini adalah bagian penting
for i in range(1, 7):
    # Hanya aturan yang aktif (alpha > 0) yang dihitung
    if alpha[i] > 0:
        numerator += (alpha[i] * z[i])
        denominator += alpha[i]

if denominator == 0:
    final_produksi = 0.0 # Tidak ada aturan yang aktif
else:
    final_produksi = numerator / denominator

print("--- Hasil Defuzzifikasi (Weighted Average) ---")
print(f"Total (α * z) = {numerator:.4f}   <-- Ini yang seharusnya")
print(f"Total α = {denominator:.4f}")
print(f"Produksi (Z) = {numerator:.4f} / {denominator:.4f} = {final_produksi:.4f}\n")

print("=========================================================")
print(f"Jadi, jumlah kemasan yang harus diproduksi adalah {final_produksi:.0f} kemasan.")
print("=========================================================")
```

```python
# --- Himpunan Fuzzy Permintaan (Sesuai UTS) ---

# 1. Tentukan titik-titik (point) grafiknya
x_permintaan_turun = [0, 1000, 3000, 4000]
y_permintaan_turun = [1, 1, 0, 0]

x_permintaan_naik = [0, 1000, 3000, 4000]
y_permintaan_naik = [0, 0, 1, 1]

# 2. Ambil nilai X dan hasil fuzzifikasi-nya
x_input = 2069
mu_turun_val = 0.4655
mu_naik_val = 0.5345

# 3. Plot grafiknya
plt.figure(figsize=(8, 5)) # Atur ukuran gambar
plt.plot(x_permintaan_turun, y_permintaan_turun, label='Turun', color='C0')
plt.plot(x_permintaan_naik, y_permintaan_naik, label='Naik', color='C1')

# 4. Plot garis fuzifikasi (yang putus-putus)
# Garis vertikal dari X
plt.plot([x_input, x_input], [0, max(mu_turun_val, mu_naik_val)], 'k--', alpha=0.8)

# Garis horizontal ke μ_turun
plt.plot([0, x_input], [mu_turun_val, mu_turun_val], 'C0--', alpha=0.8)
plt.plot(x_input, mu_turun_val, 'o', color='C0') # Titik di μ_turun

# Garis horizontal ke μ_naik
plt.plot([0, x_input], [mu_naik_val, mu_naik_val], 'C1--', alpha=0.8)
plt.plot(x_input, mu_naik_val, 'o', color='C1') # Titik di μ_naik

# 5. Keterangan
plt.title('Fuzifikasi Permintaan (X = 2069)', fontsize=14)
plt.xlabel('Jumlah Permintaan (Kemasan)', fontsize=12)
plt.ylabel('Derajat Keanggotaan (μ)', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
